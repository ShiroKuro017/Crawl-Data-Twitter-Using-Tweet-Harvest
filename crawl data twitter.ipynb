Crawl data twitter menggunakan Tweet-Harvest
===


`filename = 'FERRY_IRWANDI.csv'`

filename merupakan layout penamaan [filename].csv

`search_keyword = 'ferry irwandi since:2025-01-01 until:2025-05-01 lang:id'`

merupakan keyword pencarian di Twitter / X dengan layout :
[Topik yang dicari] [batas tanggal awal data yang dicari] [batas tanggal akhir data yang dicari] [bahasa dari topik yang sedang di cari]

`limit = 100`

merupakan jumlah limit pencarian yang akan di ambil (kadang bisa lebih beberapa dari limit) 

`!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}`

script untuk menjalankan sistem tweet-harvest yang berbasis Nodejs


```python
# Crawl Data

filename = 'FERRY_IRWANDI.csv'
search_keyword = 'ferry irwandi since:2025-01-01 until:2025-05-01 lang:id'
limit = 100

!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}
```

Menampilkan data .CSV menjadi DataFrame
===

```python
import pandas as pd

# jelaskan lokasi file ke CSV file
file_path = f"tweets-data/{filename}"

# mambaca CSV file ke pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

# menampilkan DataFrame
display(df)
```

Crosscek jumlah data yang di dapat
===

```python
# Cek jumlah data yang didapatkan

num_tweets = len(df)
print(f"Jumlah tweet dalam dataframe adalah {num_tweets}.")
```
